```python
# Importing Important Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading data
df = pd.read_csv("heart_failure_clinical_records.csv")
df.head()
```

```python
# Function For statistical depth of the data
def statistics(data):
    """
    Calculates various statistics for the dataset.
    """
    statistics = {
        'Median': data.median(),
        'Skewness': data.skew(),
        'Kurtosis': data.kurt(),
        'Description': data.describe()
    }
    return statistics
```

```python
statistics(df)
```

```python
# Function For plotting a bar plot
def bar():
    """
    Function to create a bar chart showing the Count of patients with Diabetes, Blood Pressure, and Gender.
    """
    # Creating subplots
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        
    # Plotting diabetes count
    diabetes_counts = df['diabetes'].value_counts().sort_index()
    diabetes_counts.plot(kind='bar', ax=axes[0], color='blue')
    axes[0].set_title('Diabetes Count')
    axes[0].set_xlabel('Diabetes')
    axes[0].set_ylabel('Count')
    axes[0].set_xticklabels(['Negative (0)', 'Positive (1)'])
    
    # Plotting high blood pressure count
    hbp_counts = df['high_blood_pressure'].value_counts().sort_index()
    hbp_counts.plot(kind='bar', ax=axes[1], color='green')
    axes[1].set_title('High Blood Pressure Count')
    axes[1].set_xlabel('High Blood Pressure')
    axes[1].set_xticklabels(['Negative (0)', 'Positive (1)'])
    
    # Plotting sex count
    sex_counts = df['sex'].value_counts().sort_index()
    sex_counts.plot(kind='bar', ax=axes[2], color='red')
    axes[2].set_title('Sex Count')
    axes[2].set_xlabel('Sex')
    axes[2].set_xticklabels(['Female (0)', 'Male (1)'])
    
    plt.tight_layout()
    plt.show()
```

```python
bar()
```

```python
# Function For plotting a scatter plot
def scatter():
    """
    Level of the CPK enzyme in the blood (mcg/L) vs Platelets.
    """
    plt.scatter(df['platelets'], df['creatinine_phosphokinase'], c=df['DEATH_EVENT'], cmap='coolwarm')
    plt.colorbar(label='0 = Dead  1 = Alive', ticks=[df['DEATH_EVENT'].min(), df['DEATH_EVENT'].max()])
    plt.title('Level of the CPK enzyme in the blood (mcg/L) vs Platelets')
    plt.xlabel(' Platelets in the blood (kiloplatelets/mL)')
    plt.ylabel('Level of the CPK enzyme in the blood (mcg/L)')
    plt.show()
```

```python
scatter()
```

```python
# Function For plotting a Heatmap from a Correlation Matrix
def heatmap():
    """
    Function to create a correlation matrix heatmap.
    """
    correlation_matrix = df.corr()
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True)
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
heatmap()
```
